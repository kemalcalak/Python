```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_airbnb = pd.read_csv('/content/AB_NYC_2019.csv')
```

```python
df_airbnb.head()
```

```python
df_airbnb.info()
```

```python
df_airbnb.columns
```

```python
df_airbnb.drop(['id','host_id','host_name'], axis=1,inplace=True)
```

```python
sns.set_theme(style="darkgrid")
```

```python
df_airbnb.isnull().sum()
```

```python
df_airbnb.describe()
```

```python
df_airbnb.fillna({"reviews_per_month":0, "name":0, "last_review":0}, inplace=True)
df_airbnb.isnull().sum()
```

```python
sns.histplot(df_airbnb["price"],kde=True, bins=50, )
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
df_airbnb['neighbourhood_group'].value_counts().plot.barh().set_title("Group frequency")
```

```python
df_airbnb.groupby("neighbourhood_group").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(8,8))
```

```python
df_airbnb.groupby('neighbourhood_group').agg({'price':'mean'}).plot.bar(alpha=1,color='black').set_title("Group price average")
```

```python
df_airbnb.groupby('neighbourhood_group').agg({'price':'mean','availability_365':'mean'}).plot.bar(alpha=0.7).set_title("Group price average")
plt.legend(['Price','Availability'])
```

```python
df_airbnb_1 = df_airbnb.groupby(by=['neighbourhood_group', 'room_type']).agg({'price':'mean'})
df_airbnb_1.reset_index(inplace=True)
print(df_airbnb_1)
```

```python
sns.barplot(x=df_airbnb_1['neighbourhood_group'],y=df_airbnb_1['price'],hue=df_airbnb_1['room_type'])
plt.title("Neighborhood - Price - Room")
```

```python
sns.scatterplot( x= df_airbnb['price'], y=df_airbnb['availability_365'])
```

```python
Group_price_average= df_airbnb.groupby('neighbourhood_group').agg({'price':'mean'})
sns.scatterplot( x= df_airbnb['price'], y=df_airbnb['minimum_nights'])
```

```python
sns.jointplot(x="price", y=df_airbnb["neighbourhood_group"], data=df_airbnb)
```

```python
sns.catplot(x="neighbourhood_group", y="price", data=df_airbnb)
plt.title("Neighborhood-Price")
```

```python
sns.catplot(data=df_airbnb, x="neighbourhood_group", y="price", kind= 'strip' ,hue="room_type")
plt.title("Price average")
```

```python
sns.catplot(x="neighbourhood_group", y="price", hue="room_type", kind="point",data=df_airbnb)
plt.title("Price average")
```

```python
plt.title("Correlation Matrix")
sns.heatmap(df_airbnb.corr(), annot=True);
```
