# Tensörler
Bu notebook'da tensör kavramı üzerine uygulamalar yapılacak. Doğrusal cebir, fizik, mekanizma tekniği gibi derslerde; vektör, matris, skalar kavramları sıklıkla duymuş olsak da tensör kavramıyla pek karşılaşmamışızdır. Şuana kadar öğrendiğimiz herşey aslında bir tensördür.  

Tensörler, en sade tanımıyla, veri taşıyan yapılardır. Yapay zeka ve yapay zekanın alt disiplinlerinde verileri işlemek adına farklı yöntemler uygularız. Verileri işleyebilmemiz için öncelikle bu verileri hafızada saklamamız ve istediğimizde geri çağrabilmemiz gerekir. İşte tüm bunları tensörler aracılığıyla yaparız.

tensör.JPG


Yukarıdaki görselden de anlaşılacağı üzere tensörler eksen sayılarına göre sınıflandırılırlar:


*   0 Boyutlu Tensör: Skaler sayılar bu sınıfa girer. Eksen sayıları 0'dır.
*   1 Boyutlu Tensör: Vektörler bu sınıfa girer. Eksen sayıları 1'dir.
*   2 Boyutlu Tensör: Matrisler bu sınıfa girer. Eksen sayıları 2'dir.
*   3 ve Daha Fazla Boyutlu Tensör


Tensörler hakkında bilmemiz gereken 2 önemli kavram var:

1.   Eksen Sayısı
2.   Şekil (Shape)



Eksen sayısının ne demek olduğunu az önce inceledik. Bir tensörün şekli, o tensör içeriğiyle ilgili bilgiler verir. Örneğin,


1.   Diksıra Sayısı (Sütun)
2.   Yataysıra Sayısı (Satır)
3.   Örnek Sayısı
4.   Kanal Verisi

Bunlar sadece bazı örneklerdir. Konuyu daha net anlamak için aşağıdaki örnekleri inceleyebilirsiniz.


Şimdi sırasıyla bu tensörleri inceleyelim. Öncelikle numpy kütüphanesi ile temel işlemler yaparak tensör mantığını anlamaya çalışalım.

```python
import numpy as np
```

Aşağıdaki örnekleri incelerken bilmeniz gereken 3 önemli metot var:


1.   np.array( ): Dizi oluşturan metottur.
2.   np.ndim:     Eksen sayısını veren metottur.
3.   np.shape:    Şekli veren metottur.




**0B Tensörler (Skaler)**

```python
x = np.array(7)
x.ndim
x.shape
```

```python
y = np.array(4)
y.ndim
y.shape
```

**1B Tensörler (Vektör)**

```python
x = np.array([5,8,9,7,7])
x.ndim
x.shape
```

```python
y = np.array([2,8,7])
y.ndim
y.shape
```

**2B Tensör (Matris)**

```python
x = np.array([[5,8,9,7],
             [7,6,42,1]])
x.ndim
x.shape
```

```python
y = np.array([[1,2],
              [6,8],
              [-9,7]])
y.ndim
y.shape
```

**3B Tensörler**

```python
x = np.array([[[5,8,9,7],[7,6,42,1]],
              [[5,8,9,7],[7,6,42,1]],
              [[5,8,9,7],[7,6,42,1]]])
x.ndim
x.shape
```
