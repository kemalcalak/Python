```python
# library
from keras.preprocessing.image import load_img
```

```python
# reading/loading image
img = load_img("datasets/yoda.jpg")

print(img, end="\n\n")

print("Type:",type(img))
print("Format:",img.format)
print("Size:",img.size)
print("Mode:",img.mode)
```

```python
# displaying image
# pip install matplotlib
import matplotlib.pyplot as plt
img = load_img("datasets/yoda.jpg")
img1 = load_img("datasets/clone.jpg")

plt.imshow(img)
plt.show()

plt.imshow(img1)
plt.show()
```

```python
# PIL --> numpy array
from keras.preprocessing.image import load_img, img_to_array, array_to_img

img = load_img("datasets/yoda.jpg")
print("Type:",type(img))

img = img_to_array(img)
print("Type:",type(img))


# numpy array --> PIL
img = array_to_img(img)
print("Type:",type(img))
```
